## Immediate Overview

- `df.describe()` — stats for numeric cols

## Missing Values

- `df.isnull().sum()` — count NaNs per column
- `df.isnull().mean() * 100` — % missing per column
- `df.isnull().sum(axis=1)` — NaNs per row
- `df[df.isnull().any(axis=1)]` — rows with any NaN
- `df.dropna(thresh=len(df)*0.5, axis=1)` — drop cols >50% missing

```python
df.isnull().sum()
```

```python
df.isnull().mean() * 100
```

```python
df.isnull().sum(axis=1)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(thresh=len(df)*0.5, axis=1)
```

## Distributions & Value Counts

- `df['col'].value_counts()` — frequency table
- `df['col'].value_counts(normalize=True)` — as proportions
- `df['col'].nunique()` — count of unique values
- `df['col'].unique()` — see what they are

```python
df['Survived'].value_counts(normalize=True)
```
