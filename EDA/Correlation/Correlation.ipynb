```python
import pandas as pd
titanic_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(titanic_url)
```

## Correlations & Relationships

- `df.corr(numeric_only=True)` — full correlation matrix across all numeric columns
- `df.corr()['target'].sort_values()` — correlation of every column against your target, ranked
- `df.groupby('cat_col')['num_col'].mean()` — mean of a numeric column grouped by a category

```python
df.corr(numeric_only=True)
```

```python
df.corr(numeric_only=True)['Survived'].sort_values(ascending=False)
```

```python
df.groupby('Sex').mean(numeric_only=True)
```

```python
import plotly.express as px

corr = df.corr(numeric_only=True)

fig = px.imshow(corr,
                text_auto='.2f',
                color_continuous_scale='RdBu_r',
                zmin=-1, zmax=1,
                title='Correlation Matrix')
fig.show()
```
