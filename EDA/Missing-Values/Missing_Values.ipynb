Missing Values

```python
import pandas as pd
```

```python
titanic_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(titanic_url)
```

## Missing Values

- `df.isnull().sum()` — count NaNs per column
- `df.isnull().mean() * 100` — % missing per column
- `df.isnull().sum(axis=1)` — NaNs per row
- `df[df.isnull().any(axis=1)]` — rows with any NaN
- `df.dropna(thresh=len(df)*0.5, axis=1)` — drop cols >50% missing

```python
df.isnull().sum()
```

```python
df.isnull().mean() * 100
```

```python
df.isnull().sum(axis=1)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(thresh=len(df)*0.5, axis=1)
```

```python
import plotly.express as px

missing = df.isnull().mean() * 100
missing = missing[missing > 0].sort_values(ascending=False).reset_index()
missing.columns = ['column', 'pct_missing']

fig = px.bar(missing, x='column', y='pct_missing',
             text='pct_missing',
             color='column',
             title='% Missing Values per Column',
             labels={'pct_missing': '% Missing'})
fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.show()
```

```python
import plotly.express as px

fig = px.imshow(df.isnull(),
                color_continuous_scale=['white', 'red'],
                title='Missing Value Map (red = missing)',
                labels={'color': 'Is Missing'})
fig.update_layout(coloraxis_showscale=False)
fig.show()
```

```python
import plotly.graph_objects as go

missing_df = pd.DataFrame({
    'column': df.columns,
    'missing_count': df.isnull().sum().values,
    'pct_missing': (df.isnull().mean() * 100).round(2).values
}).sort_values('pct_missing', ascending=False)

fig = go.Figure(data=[go.Table(
    header=dict(values=['Column', 'Missing Count', '% Missing'],
                fill_color='steelblue', font_color='white'),
    cells=dict(values=[missing_df.column,
                       missing_df.missing_count,
                       missing_df.pct_missing])
)])
fig.update_layout(title='Missing Value Summary')
fig.show()
```
