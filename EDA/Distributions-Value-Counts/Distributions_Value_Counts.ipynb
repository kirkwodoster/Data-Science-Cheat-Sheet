```python
import pandas as pd
titanic_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(titanic_url)
```

```python
summary = pd.DataFrame({
    'dtype': df.dtypes,
    'unique_count': df.nunique(),
    'top_value': [df[col].value_counts().index[0] for col in df.columns],
    'top_freq': [df[col].value_counts().iloc[0] for col in df.columns],
    'top_pct': [(df[col].value_counts(normalize=True).iloc[0] * 100).round(2) for col in df.columns]
})
summary
```
